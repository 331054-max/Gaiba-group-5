```python
# Single runnable cell for Jupyter. Fixes `df` undefined and adds robust fallback logic.
# Put 'Gaiba resume dataset.csv' in the same folder as this notebook, or change CSV_PATH.

USE_LLM = False   # Set True only if you have groq client & GROQ_API_KEY in env
CSV_PATH = "Gaiba resume dataset.csv"  # change if needed
N_ROWS = 20  # how many rows to test/process

# -----------------------
# Imports
# -----------------------
from tqdm import tqdm
import pandas as pd
import numpy as np
import re, json, os

# Optional Groq client import (only if USE_LLM True)
client = None
if USE_LLM:
    try:
        from groq import Groq
        if not os.environ.get("GROQ_API_KEY"):
            raise RuntimeError("Set GROQ_API_KEY in environment before using USE_LLM=True")
        client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
    except Exception as e:
        client = None
        print("Groq client not available or API key missing. Falling back to heuristic. Error:", e)
        USE_LLM = False

# -----------------------
# Load dataset into df
# -----------------------
try:
    df = pd.read_csv(CSV_PATH)
    print(f"Loaded {len(df)} rows from '{CSV_PATH}'")
except Exception as e:
    raise FileNotFoundError(f"Could not load '{CSV_PATH}' - error: {e}")

# -----------------------
# Example JD dict -> string (so both formats work)
# -----------------------
job_description = {
    "Job Title": "AI Researcher",
    "Overview": "We are seeking an experienced AI Researcher to join our innovation lab. The ideal candidate will have a strong background in TensorFlow, NLP, and PyTorch, with proven expertise in designing advanced AI solutions. You will collaborate with cross-functional teams to create, test, and deploy cutting-edge AI systems.",
    "Requirements": [
        "7–10 years of experience in AI/ML research or applied data science.",
        "Proficiency in Python, TensorFlow, and PyTorch.",
        "Strong mathematical/statistical foundation.",
        "Bachelor’s or higher degree in Computer Science, AI, or related field."
    ]
}
# convert to text if a dict
if isinstance(job_description, dict):
    jd_text = "\n".join([str(v) for v in job_description.values()])
else:
    jd_text = str(job_description)

# -----------------------
# Helper functions
# -----------------------
def safe_get(row, keys, default=""):
    """Try multiple possible column names in row; return first found."""
    for k in keys:
        if k in row and pd.notna(row[k]):
            return row[k]
    return default

def build_resume_text(row):
    # Try common column names; fallback to empty
    return f"""Name: {safe_get(row, ['Name','name','candidate','candidate_name'])}
Job Role: {safe_get(row, ['Job Role','job_role','title','designation'])}
Experience: {safe_get(row, ['Experience (Years)','experience_years','experience','yoe'])}
Skills: {safe_get(row, ['Skills','skills','skillset'])}
Education: {safe_get(row, ['Education','education','highest_education'])}
Certifications: {safe_get(row, ['Certifications','certifications'])}
Projects Count: {safe_get(row, ['Projects Count','projects_count','projects'])}
Salary Expectation: {safe_get(row, ['Salary Expectation ($)','salary_expectation','$salary'])}
AI Score: {safe_get(row, ['AI Score (0-100)','ai_score','ai_score_0_100'])}
"""

# Simple JD parser (extract skills tokens and experience min/max)
def parse_jd(jd_text):
    txt = jd_text.lower()
    # years min/max
    m = re.search(r'(\d{1,2})\s*(?:[-–to]{1,3})\s*(\d{1,2})\s*years', txt)
    if m:
        jd_min, jd_max = int(m.group(1)), int(m.group(2))
    else:
        m2 = re.search(r'(\d{1,2})\s*\+\s*years', txt)
        if m2:
            jd_min = int(m2.group(1)); jd_max = jd_min + 2
        else:
            jd_min, jd_max = 2, 4
    # skills: simple tokens found from a shortlist
    common = ['python','tensorflow','pytorch','nlp','deep','machine','ml','pandas','numpy','sql','aws','gcp','azure']
    skills = set([t for t in common if t in txt])
    return jd_min, jd_max, skills

JD_MIN, JD_MAX, JD_SKILLS = parse_jd(jd_text)

# Heuristic scorer (returns parsed dict like the LLM expected format)
def heuristic_score(resume_text, row):
    # compute Experience score (0-10)
    exp_raw = safe_get(row, ['Experience (Years)','experience_years','experience','yoe'])
    try:
        exp_val = float(str(exp_raw).replace('+','').strip())
    except:
        exp_val = None
    if exp_val is None:
        exp_score = 5
    else:
        # scale: 0->0, JD_MIN..JD_MAX -> 6..8, >=JD_MAX+3 ->10
        if exp_val >= JD_MAX + 3:
            exp_score = 10
        elif exp_val >= JD_MAX:
            exp_score = 8
        elif exp_val >= JD_MIN:
            exp_score = 6
        else:
            exp_score = max(0, int( (exp_val / max(1, JD_MIN)) * 6 ))

    # skills match (0-10)
    skills_cell = str(safe_get(row, ['Skills','skills','skillset'])).lower()
    have = set([re.sub(r'[^a-z0-9]+','',s) for s in re.split(r'[;,/|\s]+', skills_cell) if s.strip()])
    if JD_SKILLS:
        matches = len(JD_SKILLS & have)
        skill_score = min(10, int(round(matches / max(1,len(JD_SKILLS)) * 10)))
    else:
        skill_score = 5

    # projects (approx: based on Projects Count)
    proj_raw = safe_get(row, ['Projects Count','projects_count','projects'])
    try:
        proj_val = int(re.sub(r'[^0-9]', '', str(proj_raw))) if pd.notna(proj_raw) and str(proj_raw).strip()!="" else 0
    except:
        proj_val = 0
    proj_score = min(10, proj_val if proj_val<=10 else 10)

    # certs/tools (simple check for known tool tokens)
    certs_cell = str(safe_get(row, ['Certifications','certifications','tools'])).lower()
    certs_have = len([t for t in ['aws','gcp','azure','tensorflow','pytorch','spark'] if t in certs_cell])
    cert_score = min(10, certs_have * 3)

    # education/domain (naive)
    edu = str(safe_get(row, ['Education','education','highest_education'])).lower()
    edu_score = 5
    if any(x in edu for x in ['phd','doctorate']): edu_score = 10
    elif any(x in edu for x in ['master','m.sc','msc','mba','mtech']): edu_score = 8
    elif any(x in edu for x in ['bachelor','b.sc','btech','b.e']): edu_score = 6

    total = exp_score + skill_score + proj_score + cert_score + edu_score
    # summarise
    summary = f"Exp {exp_score}/10, Skills {skill_score}/10, Projects {proj_score}/10, Certs {cert_score}/10, Edu {edu_score}/10"
    shortlist = "Yes" if (skill_score>=4 and exp_score>=4 and edu_score>=4) else "No"
    return {
        "Experience": exp_score,
        "Skills": skill_score,
        "Projects": proj_score,
        "Certifications_Tools": cert_score,
        "Education_Domain": edu_score,
        "Total_Score": total,
        "Summary": summary,
        "Shortlist": shortlist
    }

# -----------------------
# NEW: Attrition risk calculator (uses available fields in Gaiba dataset if present)
# -----------------------
def compute_attrition_risk(row):
    """
    Heuristic attrition risk:
    - Looks for columns like 'Num Jobs', 'Num Job Changes', 'Job Change Rate', 'Avg Tenure Months' in the row.
    - Falls back to computing job_change_rate = num_jobs / experience_years if possible.
    - Rules:
       High if job_change_rate > 1.0 OR avg_tenure_months < 12 OR num_jobs >= 5
       Medium if job_change_rate > 0.5 OR 12 <= avg_tenure_months < 24 OR num_jobs >= 3
       Else Low
    """
    # try to read candidate experience years
    exp_raw = safe_get(row, ['Experience (Years)','experience_years','experience','yoe'])
    try:
        exp_years = float(str(exp_raw).replace('+','').strip()) if pd.notna(exp_raw) and str(exp_raw).strip()!="" else None
    except:
        exp_years = None

    # look for numeric columns that indicate number of jobs / changes
    num_jobs = None
    for c in ['Num Jobs','Num_Jobs','num_jobs','Number of Jobs','Num Job Changes','NumJobChanges','job_changes','num_job_changes']:
        if c in row and pd.notna(row[c]):
            try:
                num_jobs = int(re.sub(r'[^0-9]', '', str(row[c])))
                break
            except:
                num_jobs = None

    # look for avg tenure months
    avg_tenure_months = None
    for c in ['Avg Tenure Months','Avg_Tenure_Months','avg_tenure_months','average_tenure_months','avg_tenure']:
        if c in row and pd.notna(row[c]):
            try:
                avg_tenure_months = float(re.sub(r'[^0-9.]', '', str(row[c])))
                break
            except:
                avg_tenure_months = None

    # look for explicit job_change_rate_per_year
    job_change_rate = None
    for c in ['Job Change Rate','job_change_rate_per_year','job_change_rate','change_rate']:
        if c in row and pd.notna(row[c]):
            try:
                job_change_rate = float(re.sub(r'[^0-9.]', '', str(row[c])))
                break
            except:
                job_change_rate = None

    # fallback: compute job_change_rate if num_jobs and exp_years present
    if job_change_rate is None and num_jobs is not None and exp_years and exp_years > 0:
        job_change_rate = num_jobs / exp_years

    # Salary-based heuristic: high salary expectation with low experience can indicate flight risk
    salary_raw = safe_get(row, ['Salary Expectation ($)','salary_expectation','$salary','Expected Salary','Salary'])
    salary_val = None
    try:
        if pd.notna(salary_raw) and str(salary_raw).strip()!="":
            salary_val = float(re.sub(r'[^\d.]','', str(salary_raw)))
    except:
        salary_val = None

    # Now determine risk
    reasons = []
    risk = "Low"
    try:
        if (job_change_rate is not None and job_change_rate > 1.0) or (avg_tenure_months is not None and avg_tenure_months < 12) or (num_jobs is not None and num_jobs >= 5):
            risk = "High"
            if job_change_rate is not None and job_change_rate > 1.0:
                reasons.append(f"High job_change_rate {job_change_rate:.2f}/yr")
            if avg_tenure_months is not None and avg_tenure_months < 12:
                reasons.append(f"Low avg tenure {avg_tenure_months:.1f} months")
            if num_jobs is not None and num_jobs >= 5:
                reasons.append(f"{num_jobs} jobs recorded")
        elif (job_change_rate is not None and job_change_rate > 0.5) or (avg_tenure_months is not None and 12 <= avg_tenure_months < 24) or (num_jobs is not None and num_jobs >= 3):
            risk = "Medium"
            if job_change_rate is not None and job_change_rate > 0.5:
                reasons.append(f"Moderate job_change_rate {job_change_rate:.2f}/yr")
            if avg_tenure_months is not None and 12 <= avg_tenure_months < 24:
                reasons.append(f"Avg tenure {avg_tenure_months:.1f} months")
            if num_jobs is not None and num_jobs >= 3:
                reasons.append(f"{num_jobs} jobs recorded")
        else:
            risk = "Low"
            reasons.append("Stable history by heuristics")
        # salary factor: raise one level if salary high relative to experience
        if salary_val is not None and exp_years is not None:
            if salary_val > 150000 and exp_years < 5 and risk == "Low":
                risk = "Medium"; reasons.append(f"High salary ${salary_val:,.0f} for {exp_years} yrs exp")
            elif salary_val > 150000 and exp_years < 5 and risk == "Medium":
                risk = "High"; reasons.append(f"Very high salary ${salary_val:,.0f} for {exp_years} yrs exp")
    except Exception as e:
        # fallback safe
        risk = "Unknown"
        reasons.append(f"Error computing risk: {e}")

    return risk, "; ".join(reasons)

# -----------------------
# Main loop: iterate rows and either call LLM or heuristic
# -----------------------
matched_results = []

rows_to_process = min(N_ROWS, len(df))
for idx, row in tqdm(df.head(rows_to_process).iterrows(), total=rows_to_process, desc="Processing"):
    resume_text = build_resume_text(row)
    parsed = {}
    if USE_LLM and client is not None:
        # build prompt (compact)
        prompt = f"You are an expert recruiter. JD: {jd_text}\n\nCandidate:\n{resume_text}\n\nReturn JSON with keys: Experience, Skills, Projects, Certifications_Tools, Education_Domain, Total_Score, Summary, Shortlist"
        try:
            resp = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[{"role":"system","content":"You are a professional resume screening AI assistant."},
                          {"role":"user","content": prompt}],
                temperature=0.0
            )
            raw = resp.choices[0].message.content.strip()
            # try JSON parse
            try:
                parsed = json.loads(raw)
            except:
                # try extract JSON substring
                m = re.search(r"(\{.*\})", raw, flags=re.S)
                if m:
                    parsed = json.loads(m.group(1))
                else:
                    parsed = {"Summary": raw, "Shortlist": "No"}
        except Exception as e:
            print("LLM call error:", e)
            parsed = heuristic_score(resume_text, row)
    else:
        parsed = heuristic_score(resume_text, row)

    # -----------------------
    # NEW: compute attrition risk and attach
    # -----------------------
    risk_label, risk_reasons = compute_attrition_risk(row)

    matched_results.append({
        "Resume_Text": resume_text,
        "Experience_score": parsed.get("Experience"),
        "Skills_score": parsed.get("Skills"),
        "Projects_score": parsed.get("Projects"),
        "Certifications_Tools_score": parsed.get("Certifications_Tools"),
        "Education_score": parsed.get("Education_Domain"),
        "Total_Score": parsed.get("Total_Score"),
        "Summary": parsed.get("Summary", ""),
        "Shortlist": parsed.get("Shortlist", "No"),
        "Attrition_Risk": risk_label,            # <--- added field
        "Attrition_Risk_Reasons": risk_reasons   # optional explanatory reasons
    })

# Build final DataFrame
final_df = pd.DataFrame(matched_results)
print("Processed rows:", len(final_df))
final_df.head(20)

 
```

```python
# Paste this whole cell into Jupyter and run.
from tqdm import tqdm
import json, os, re
import pandas as pd
import numpy as np

# Optional Groq LLM usage (set True if you want to call Groq and have API key)
USE_LLM = False

try:
    if USE_LLM:
        from groq import Groq
except Exception:
    if USE_LLM:
        print("Groq not available - falling back to USE_LLM = False")
        USE_LLM = False

# -------------------------
# Helper utilities
# -------------------------
DEGREE_ORDER = ["highschool","diploma","bachelor","master","phd"]
DEGREE_MAP = {
    r"\b(high\s*school|12th|hsc|ssc|secondary)\b": "highschool",
    r"\b(diploma|polytechnic)\b": "diploma",
    r"\b(bachelor|b\.?tech|bsc|ba|b\.?com|bca)\b": "bachelor",
    r"\b(master|m\.?tech|msc|ma|mba|mca)\b": "master",
    r"\b(ph\.?d|doctorate|dphil)\b": "phd",
}
ROLE_ORDER = ["intern","junior","mid","senior","lead","manager","director","executive"]
ROLE_HINTS = {
    "intern":"intern|trainee",
    "junior":"junior|\\bjr\\b|associate",
    "mid":"analyst|engineer|consultant|specialist",
    "senior":"senior|\\bsr\\b|staff|principal",
    "lead":"\\blead\\b|team lead|tech lead",
    "manager":"manager|\\bmgr\\b|management|head of",
    "director":"director|\\bhead\\b",
    "executive":"\\bvp\\b|vice president|chief|cxo|cto|cfo|ceo"
}

def normalize_degree(text):
    if not isinstance(text,str): return None
    t = text.lower()
    for pat, val in DEGREE_MAP.items():
        if re.search(pat, t): return val
    if "ph" in t and "d" in t: return "phd"
    if "mba" in t or "master" in t: return "master"
    if "bachelor" in t or "b." in t: return "bachelor"
    if "diploma" in t: return "diploma"
    if "school" in t: return "highschool"
    return None

def infer_role_level(title):
    if not isinstance(title,str) or not title.strip(): return None
    t = title.lower()
    for level in ROLE_ORDER[::-1]:
        pat = ROLE_HINTS.get(level)
        if pat and re.search(pat, t): return level
    if "senior" in t or " sr " in t: return "senior"
    if "manager" in t: return "manager"
    return "mid"

def idx(lst, val):
    try: return lst.index(val)
    except: return None

def extract_jd_requirements(text):
    jd_min, jd_max = 2, 4
    jd_degree = "bachelor"
    jd_role = "junior"
    if not isinstance(text,str): return jd_min, jd_max, jd_degree, jd_role
    txt = text.lower()
    m = re.search(r'(\d{1,2})\s*(?:[-–to]{1,3})\s*(\d{1,2})\s*years', txt)
    if m:
        jd_min, jd_max = int(m.group(1)), int(m.group(2))
    else:
        m2 = re.search(r'(\d{1,2})\s*\+\s*years', txt)
        if m2:
            jd_min = int(m2.group(1)); jd_max = jd_min + 2
    for pat,deg in DEGREE_MAP.items():
        if re.search(pat, txt):
            jd_degree = deg
            break
    for lvl in ROLE_ORDER[::-1]:
        if re.search(r'\b'+re.escape(lvl)+r'\b', txt):
            jd_role = lvl
            break
    return jd_min, jd_max, jd_degree, jd_role

def compute_overqualification(exp_years, role_level, edu_level, jd_min, jd_max, jd_role_idx, jd_edu_idx):
    reasons = []
    over = False
    # experience-based check
    try:
        if exp_years is not None:
            ev = float(exp_years)
            if jd_max is not None and ev >= (jd_max + 3):
                over = True
                reasons.append(f"{ev:.1f} yrs vs JD {jd_min}-{jd_max}")
    except:
        pass
    # seniority gap
    try:
        if role_level is not None and jd_role_idx is not None:
            cand_idx = idx(ROLE_ORDER, role_level)
            if cand_idx is not None and cand_idx - jd_role_idx >= 2:
                over = True
                reasons.append(f"title '{role_level}' above JD '{ROLE_ORDER[jd_role_idx]}'")
    except:
        pass
    # education gap
    try:
        if edu_level is not None and jd_edu_idx is not None:
            edu_idx = idx(DEGREE_ORDER, edu_level)
            if edu_idx is not None and edu_idx - jd_edu_idx >= 2:
                over = True
                reasons.append(f"education '{edu_level}' >> JD '{DEGREE_ORDER[jd_edu_idx]}'")
    except:
        pass
    return over, "; ".join(reasons) if reasons else ""

def compute_education_fit(edu_level, jd_edu_idx):
    if jd_edu_idx is None: return "Unknown", "No JD degree requirement"
    if edu_level is None: return "Underqualified", "No detectable degree"
    edu_idx = idx(DEGREE_ORDER, edu_level)
    if edu_idx is None: return "Unknown", f"Unrecognized degree {edu_level}"
    if edu_idx < jd_edu_idx: return "Underqualified", f"{edu_level} < required"
    if edu_idx == jd_edu_idx: return "Meets", f"{edu_level} meets requirement"
    return "Overqualified", f"{edu_level} > required"

# Simple heuristic parser (used when LLM not available)
def heuristic_parse(row, jd_text):
    jd_min, jd_max, jd_degree, jd_role = extract_jd_requirements(jd_text)
    # experience
    exp_candidates = []
    for cname in ['Experience (Years)','Experience','experience_years','yoe','years_of_experience']:
        if cname in row.index:
            exp_candidates.append(row.get(cname))
    exp_val = None
    for v in exp_candidates:
        try:
            if pd.notna(v) and str(v).strip() != "":
                exp_val = float(re.sub(r'[^\d\.]','', str(v)))
                break
        except:
            continue
    # skills: simple token match from JD
    jd_skills = set(re.findall(r'\b([a-z0-9\+\#]+)\b', jd_text.lower()))
    skills_text = row.get('Skills') or row.get('skills') or ""
    have = set([s.lower() for s in re.split(r'[;,/|\n]+', str(skills_text)) if s.strip()])
    skill_matches = len(jd_skills & have) if jd_skills else 0
    skill_score = min(10, int(round((skill_matches / max(1,len(jd_skills))) * 10))) if jd_skills else 5
    # projects
    try:
        proj = row.get('Projects Count') or row.get('projects_count') or 0
        proj_val = int(re.sub(r'[^\d]','', str(proj))) if pd.notna(proj) and str(proj).strip()!="" else 0
    except:
        proj_val = 0
    parsed = {
        "Experience": exp_val,
        "Skills": skill_score,
        "Projects": proj_val,
        "Certifications_Tools": None,
        "Education_Domain": row.get('Education') or row.get('education') or None,
        "Total_Score": None,
        "Summary": None,
        "Shortlist": "No"
    }
    return parsed

# -------------------------
# Main function
# -------------------------
def process_resumes(jd_text, resume_file_path, use_llm=USE_LLM):
    # create LLM client if requested
    client = None
    if use_llm:
        if not os.environ.get("GROQ_API_KEY"):
            print("GROQ_API_KEY not found. Falling back to heuristic (use_llm=False).")
            use_llm = False
        else:
            try:
                client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
            except Exception as e:
                print("Error initializing Groq client:", e)
                use_llm = False

    # load dataset
    if resume_file_path.lower().endswith('.csv'):
        df = pd.read_csv(resume_file_path)
    elif resume_file_path.lower().endswith(('.xls','.xlsx')):
        df = pd.read_excel(resume_file_path)
    else:
        raise ValueError("Unsupported file format. Provide CSV or Excel file.")

    # preserve original columns
    orig_cols = list(df.columns)

    # jd params
    jd_min, jd_max, jd_degree, jd_role = extract_jd_requirements(jd_text)
    jd_role_idx = idx(ROLE_ORDER, jd_role)
    jd_edu_idx = idx(DEGREE_ORDER, jd_degree)

    results = []
    for i, row in tqdm(df.iterrows(), total=len(df), desc="Processing resumes"):
        # build resume_text if needed for LLM
        resume_text = f"Name: {row.get('Name','')}\nTitle: {row.get('Job Role', row.get('title',''))}\nSkills: {row.get('Skills','')}\nEducation: {row.get('Education','')}\nExperience: {row.get('Experience (Years)','')}\n"
        parsed = {}
        api_error = None
        if use_llm and client is not None:
            prompt = f"You are an expert recruiter. JD: {jd_text}\n\nCandidate:\n{resume_text}\n\nReturn JSON with keys: Experience, Skills, Projects, Certifications_Tools, Education_Domain, Total_Score, Summary, Shortlist"
            try:
                resp = client.chat.completions.create(
                    model="llama-3.3-70b-versatile",
                    messages=[{"role":"system","content":"You are a professional resume screening AI assistant."},
                              {"role":"user","content":prompt}],
                    temperature=0.0
                )
                raw = resp.choices[0].message.content.strip()
                try:
                    parsed = json.loads(raw)
                except:
                    m = re.search(r'(\{.*\})', raw, flags=re.S)
                    parsed = json.loads(m.group(1)) if m else {"Summary": raw, "Shortlist":"No"}
            except Exception as e:
                api_error = str(e)
                parsed = heuristic_parse(row, jd_text)
        else:
            parsed = heuristic_parse(row, jd_text)

        # candidate education and role normalized
        cand_education_raw = parsed.get("Education_Domain") or row.get('Education') or row.get('education') or None
        cand_edu_norm = normalize_degree(str(cand_education_raw)) if cand_education_raw not in (None, np.nan) else None
        cand_role_raw = parsed.get("Job Role") or row.get('Job Role') or row.get('title') or row.get('designation') or None
        cand_role_level = infer_role_level(cand_role_raw) if cand_role_raw else None

        # parsed experience numeric fallback
        parsed_exp = parsed.get("Experience")
        try:
            parsed_exp_val = float(parsed_exp) if parsed_exp not in (None,"") else None
        except:
            parsed_exp_val = None
        if parsed_exp_val is None:
            for col_opt in ['Experience (Years)','Experience','experience_years','yoe','years_of_experience']:
                if col_opt in df.columns:
                    try:
                        v = row.get(col_opt)
                        if pd.notna(v) and str(v).strip()!="":
                            parsed_exp_val = float(re.sub(r'[^\d\.]','', str(v)))
                            break
                    except:
                        continue

        # compute overqualification and edu-fit
        is_over, over_reason = compute_overqualification(parsed_exp_val, cand_role_level, cand_edu_norm, jd_min, jd_max, jd_role_idx, jd_edu_idx)
        edu_fit_label, edu_fit_reason = compute_education_fit(cand_edu_norm, jd_edu_idx)

        # assemble output row: include all original columns then new ones
        out_row = {col: row.get(col, None) for col in orig_cols}
        out_row.update({
            "Overqualified": "Yes" if is_over else "No",
            "Overqual_Reason": over_reason,
            "Edu_to_Role_Fit": edu_fit_label,
            "Edu_Fit_Reason": edu_fit_reason,
            "API_Response_Error": api_error,
            # keep LLM parsed fields for traceability
            "Parsed_Experience": parsed.get("Experience"),
            "Parsed_Skills": parsed.get("Skills"),
            "Parsed_Projects": parsed.get("Projects"),
            "Parsed_Total_Score": parsed.get("Total_Score"),
            "Parsed_Shortlist": parsed.get("Shortlist")
        })
        results.append(out_row)

    final_df = pd.DataFrame(results)
    # display the table (first 20 rows)
    from IPython.display import display
    display(final_df.head(20))
    return final_df

# -------------------------
# Example usage (uncomment and set path)
# -------------------------
# jd_text = "7-10 years experience in AI/ML. Required: Python, TensorFlow, PyTorch. Bachelor's or higher."
# df_out = process_resumes(jd_text, "Gaiba resume dataset.csv", use_llm=False)
```

```python
# Run this entire cell in Google Colab.
# If first run in the session, uncomment the pip install line below.
# !pip -q install gradio plotly pandas numpy python-dateutil openpyxl

import os, io, re, json, math
from collections import Counter
from datetime import datetime

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import gradio as gr

# Optional: Groq LLM (only used when USE_LLM=True and GROQ_API_KEY present)
try:
    from groq import Groq
    GROQ_AVAILABLE = True
except Exception:
    GROQ_AVAILABLE = False

# Colab renderer
pio.renderers.default = "colab"

# ---------- CONFIG ----------
USE_LLM = False    # Set True only if you have GROQ_API_KEY in os.environ and want LLM scoring
MAX_PREVIEW_ROWS = 500
# ---------------------------

# ---------- Utilities ----------
def snake(s: str) -> str:
    return re.sub(r'[^a-z0-9]+','_', str(s).strip().lower()).strip('_')

def find_col(df, candidates):
    for alias in candidates:
        for c in df.columns:
            if snake(c) == snake(alias):
                return c
    for alias in candidates:
        for c in df.columns:
            if alias.lower() in c.lower():
                return c
    return None

DEGREE_ORDER = ["highschool","diploma","bachelor","master","phd"]
DEGREE_MAP = {
    r"(high\s*school|12th|hsc|ssc|secondary)":"highschool",
    r"(diploma|polytechnic)":"diploma",
    r"(bachelor|b\.?tech|bsc|ba|bcom|bca)":"bachelor",
    r"(master|m\.?tech|msc|ma|mba|mca)":"master",
    r"(ph\.?d|doctorate)":"phd"
}
ROLE_ORDER = ["intern","junior","mid","senior","lead","manager","director","executive"]
ROLE_HINTS = {
    "intern":"intern|trainee",
    "junior":"junior|\\bjr\\b|associate",
    "mid":"analyst|engineer|consultant|specialist|associate",
    "senior":"senior|\\bsr\\b|staff|principal",
    "lead":"\\blead\\b|team lead|tech lead",
    "manager":"manager|\\bmgr\\b|head of",
    "director":"director|head\\b",
    "executive":"\\bvp\\b|vice president|chief|cxo|cto|cfo|ceo"
}

def normalize_degree(text):
    if not isinstance(text,str): return None
    t = text.lower()
    for pat,val in DEGREE_MAP.items():
        if re.search(pat, t):
            return val
    if "ph" in t and "d" in t: return "phd"
    if "mba" in t or "master" in t: return "master"
    if "bachelor" in t or "b." in t: return "bachelor"
    if "diploma" in t: return "diploma"
    if "school" in t: return "highschool"
    return None

def infer_role_level(title):
    if not isinstance(title,str): return None
    t = title.lower()
    for lvl in ROLE_ORDER[::-1]:
        pat = ROLE_HINTS.get(lvl)
        if pat and re.search(pat, t):
            return lvl
    if "senior" in t or " sr " in t: return "senior"
    if "manager" in t: return "manager"
    return "mid"

def parse_skills(cell):
    if not isinstance(cell,str): return set()
    parts = re.split(r'[;,/|\n]+', cell.lower())
    parts = [p.strip() for p in parts if p.strip()]
    tokens = set()
    for p in parts:
        for sub in re.split(r'[\s/&]+', p):
            sub = sub.strip()
            if sub:
                tokens.add(sub)
    return tokens

# Robust attrition function
def compute_attrition_risk_row(row):
    # Use avg_tenure_months and job_change_rate_per_year if present; else fallback to basic heuristics
    avg_tenure = row.get('avg_tenure_months') or row.get('avg_tenure') or None
    jcr = row.get('job_change_rate_per_year') or row.get('job_change_rate') or None
    # normalize numeric
    try: avg_tenure = float(avg_tenure) if pd.notna(avg_tenure) else None
    except: avg_tenure = None
    try: jcr = float(jcr) if pd.notna(jcr) else None
    except: jcr = None

    f1 = max(0, (18 - (avg_tenure or 0)) / 18.0)
    f2 = max(0, min(1, (jcr or 0) / 1.5))
    score = 100*(0.6*f1 + 0.4*f2)
    if score >= 65:
        return "High", f"Risk score {score:.1f}"
    elif score >= 35:
        return "Medium", f"Risk score {score:.1f}"
    else:
        return "Low", f"Risk score {score:.1f}"

def compute_qualification_status(exp_years, role_level, edu_level, jd_min, jd_max, jd_edu):
    reasons = []
    status = "Good Fit"
    # normalize inputs
    try:
        exp_val = float(exp_years) if exp_years is not None and str(exp_years) != "" else None
    except:
        exp_val = None
    edu_norm = normalize_degree(edu_level) if edu_level else None
    JD_EDU_IDX = DEGREE_ORDER.index(jd_edu) if jd_edu in DEGREE_ORDER else None
    # Overqualification
    if exp_val is not None and jd_max is not None:
        if exp_val >= (jd_max + 3):
            status = "Overqualified"
            reasons.append(f"{exp_val:.1f} yrs vs JD {jd_min}-{jd_max}")
    # Seniority gap
    try:
        cand_idx = ROLE_ORDER.index(role_level) if role_level in ROLE_ORDER else None
        jd_role_idx = ROLE_ORDER.index("junior") # base JD role fallback
        if cand_idx is not None and cand_idx - jd_role_idx >= 2:
            status = "Overqualified"
            reasons.append(f"title {role_level} senior vs JD")
    except:
        pass
    # Underqualification
    if status != "Overqualified":
        if exp_val is not None and jd_min is not None:
            if exp_val < (jd_min - 1):
                status = "Underqualified"
                reasons.append(f"{exp_val:.1f} yrs < JD min {jd_min}")
        if edu_norm and JD_EDU_IDX is not None:
            try:
                edu_idx = DEGREE_ORDER.index(edu_norm)
                if edu_idx < JD_EDU_IDX:
                    status = "Underqualified"
                    reasons.append(f"edu {edu_norm} < JD {jd_edu}")
            except:
                pass
    return status, "; ".join(reasons) if reasons else "Meets JD criteria"

# Minimal JD parser
def extract_jd_requirements(jd_text):
    if not isinstance(jd_text,str):
        return 2,4,"bachelor","junior", set()
    text = jd_text.lower()
    # years
    m = re.search(r'(\d{1,2})\s*(?:[-–to]{1,3})\s*(\d{1,2})\s*years', text)
    if m:
        jd_min = int(m.group(1)); jd_max = int(m.group(2))
    else:
        m2 = re.search(r'(\d{1,2})\s*\+\s*years', text)
        if m2:
            jd_min = int(m2.group(1)); jd_max = jd_min+2
        else:
            jd_min, jd_max = 2,4
    # degree
    jd_degree = None
    for pat,deg in DEGREE_MAP.items():
        if re.search(pat, text):
            jd_degree = deg
            break
    jd_degree = jd_degree or "bachelor"
    # role
    jd_role = "junior"
    for lvl in ROLE_ORDER[::-1]:
        if re.search(r'\b'+re.escape(lvl)+r'\b', text):
            jd_role = lvl
            break
    # skills
    skills = set()
    # look for "skills:" or "requirements:"
    sm = re.search(r'(skills|requirements|must have)[:\s]*(.*)', jd_text, flags=re.I|re.S)
    if sm:
        tail = sm.group(2)[:400]
        parts = re.split(r'[,\n;/]+', tail)
        for p in parts:
            t = re.sub(r'[^a-z0-9\+\#\.-]+',' ', p.lower()).strip()
            if t:
                for token in t.split():
                    if len(token) >= 2:
                        skills.add(token)
    # fallback common tech
    for t in ['python','sql','tableau','powerbi','excel','pandas','numpy','r','aws','gcp','azure','spark']:
        if t in text:
            skills.add(t)
    return jd_min, jd_max, jd_degree, jd_role, skills

# ---------- Chart helpers (expect standardized final_df columns) ----------
def pie_from_series(series, title):
    dfc = series.reset_index()
    dfc.columns = ['category','count']
    fig = px.pie(dfc, names='category', values='count', title=title, hole=0.35)
    fig.update_traces(textposition='inside', textinfo='percent+label')
    return fig

def bar_from_counts(counts, title, x_title=None, y_title="count", horizontal=False):
    dfc = counts.reset_index()
    dfc.columns = [dfc.columns[0], 'count']
    if horizontal:
        fig = px.bar(dfc, x='count', y=dfc.columns[0], orientation='h', title=title)
    else:
        fig = px.bar(dfc, x=dfc.columns[0], y='count', title=title)
    if x_title: fig.update_layout(xaxis_title=x_title)
    fig.update_layout(yaxis_title=y_title)
    return fig

# ---------- Core processing function ----------
def process_file_and_analyze(company_name, jd_text, resume_file):
    # read file robustly
    if resume_file is None:
        return "Please upload a resume CSV/XLSX file.", None, None, None, None, None, None, None, None
    # Gradio gives an object with .name (path)
    try:
        path = resume_file.name
        if path.lower().endswith('.csv'):
            df = pd.read_csv(path)
        else:
            df = pd.read_excel(path)
    except Exception:
        # try reading bytes
        try:
            resume_file.seek(0)
            content = resume_file.read()
            df = pd.read_csv(io.StringIO(content.decode('utf-8', errors='ignore')))
        except Exception as e:
            return f"Failed to read uploaded file: {e}", None, None, None, None, None, None, None, None

    if df.empty:
        return "Uploaded dataset is empty.", None, None, None, None, None, None, None, None

    # normalize column names to snake-case keys for consistent access
    orig_cols = list(df.columns)
    rename_map = {c: snake(c) for c in orig_cols}
    df.rename(columns=rename_map, inplace=True)

    # canonical columns (best-effort)
    cand_col = find_col(df, ['name','candidate','candidate_name','full_name','applicant']) or list(df.columns)[0]
    title_col = find_col(df, ['latest_title','title','role','designation','position'])
    edu_col = find_col(df, ['highest_education','education','qualification','degree'])
    exp_col = find_col(df, ['total_experience_years','experience_years','years_of_experience','yoe','experience'])
    skills_col = find_col(df, ['skills','skillset','key_skills','technical_skills'])
    tenure_col = find_col(df, ['avg_tenure_months','avg_tenure','average_tenure','tenure_months'])
    num_jobs_col = find_col(df, ['num_jobs','number_of_jobs','jobs_count'])

    # extract JD params
    jd_min, jd_max, jd_degree, jd_role, jd_skills = extract_jd_requirements(jd_text)

    # initialize Groq client only if required
    groq_client = None
    if USE_LLM:
        if not GROQ_AVAILABLE:
            return "Groq package not available. Either install groq or set USE_LLM=False.", None, None, None, None, None, None, None, None
        if not os.environ.get("GROQ_API_KEY"):
            return "GROQ_API_KEY not set. Set it in Colab environment or set USE_LLM=False.", None, None, None, None, None, None, None, None
        groq_client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

    processed = []
    # iterate rows safely (no tqdm in Gradio)
    for i, row in df.iterrows():
        try:
            # read raw values with canonical names
            candidate = row.get(cand_col, f"id_{i}")
            title = row.get(title_col, "") if title_col else ""
            edu_raw = row.get(edu_col, "") if edu_col else ""
            exp_raw = row.get(exp_col, "") if exp_col else None
            skills_raw = row.get(skills_col, "") if skills_col else ""
            tenure_raw = row.get(tenure_col, None) if tenure_col else None
            num_jobs = row.get(num_jobs_col, 1) if num_jobs_col else 1

            # normalizations
            edu_norm = normalize_degree(str(edu_raw)) if pd.notna(edu_raw) else None
            role_level = infer_role_level(str(title)) if pd.notna(title) else None
            skill_set = parse_skills(skills_raw)

            # numeric experience
            try:
                exp_val = float(str(exp_raw).strip()) if pd.notna(exp_raw) else None
            except:
                exp_val = None
            # avg tenure months fallback
            try:
                avg_tenure = float(tenure_raw) if pd.notna(tenure_raw) else None
            except:
                avg_tenure = None
            if avg_tenure is None and exp_val is not None:
                # estimate average tenure months = total exp years * 12 / num_jobs (if num_jobs present)
                try:
                    nj = float(num_jobs) if pd.notna(num_jobs) and num_jobs>0 else 1.0
                except:
                    nj = 1.0
                if exp_val is not None:
                    avg_tenure = (exp_val * 12.0) / max(1.0, nj)

            # LLM scoring (optional)
            parsed = {}
            api_error = None
            if USE_LLM and groq_client is not None:
                # build a compact prompt (we keep this conservative to avoid long calls)
                resume_text = f"Name: {candidate}\nTitle: {title}\nExperience: {exp_val}\nEducation: {edu_raw}\nSkills: {skills_raw}\n"
                jd_block = jd_text if jd_text else ""
                prompt = f"You are an expert recruiter. Rate Experience (0-10), Skills (0-10), Projects (0-10), Certifications (0-10), Education (0-10). Return JSON with keys: Experience, Skills, Projects, Certifications_Tools, Education_Domain, Total_Score, Summary, Shortlist.\n\nJob Description:\n{jd_block}\n\nResume:\n{resume_text}\n\nReturn only JSON."
                try:
                    resp = groq_client.chat.completions.create(
                        model="llama-3.1-70b-8192",
                        messages=[{"role":"system","content":"You are a professional resume screening assistant."},
                                  {"role":"user","content":prompt}],
                        temperature=0.0,
                        timeout=30
                    )
                    raw = getattr(resp.choices[0].message, "content", "") or str(resp.choices[0].message)
                    raw = raw.strip()
                    try:
                        parsed = json.loads(raw)
                    except Exception:
                        m = re.search(r'(\{.*\})', raw, flags=re.S)
                        parsed = json.loads(m.group(1)) if m else {}
                except Exception as e:
                    parsed = {}
                    api_error = f"LLM error: {e}"
            else:
                parsed = {}

            # compute attrition risk and qualification
            attrition_label, attrition_reason = compute_attrition_risk_row({'avg_tenure_months': avg_tenure, 'job_change_rate_per_year': None})
            qual_status, qual_reason = compute_qualification_status(exp_val, role_level, edu_norm, jd_min, jd_max, jd_degree)

            # skill match ratio
            skill_match_ratio = 0.0
            if jd_skills and skill_set:
                skill_match_ratio = len(jd_skills & skill_set) / max(1, len(jd_skills))

            # shortlist heuristic
            shortlist = "No"
            if (parsed.get('Shortlist') == "Yes") or (attrition_label != "High" and qual_status != "Underqualified" and skill_match_ratio >= 0.4 and qual_status != "Overqualified"):
                shortlist = "Yes"

            # summary
            matches = f"{int(skill_match_ratio*100)}% JD skills"
            tenure_summary = f"Avg tenure ~{int(avg_tenure)} months" if avg_tenure else ""
            summary_txt = parsed.get('Summary') if parsed.get('Summary') else " · ".join([matches, tenure_summary]).strip(" · ")

            # collect original row data into result
            result_row = {orig_col: row.get(snake(orig_col), None) for orig_col in orig_cols}
            # add new fields
            result_row.update({
                "summary": summary_txt,
                "attrition_risk": attrition_label,
                "attrition_reason": attrition_reason,
                "qualification_status": qual_status,
                "qualification_reason": qual_reason,
                "skill_match_ratio": round(skill_match_ratio,3),
                "shortlist": shortlist,
                "Education_Category": edu_norm,
                "Role_Level_Inferred": role_level,
                "Experience_Years_Clean": exp_val,
                "Skills_Set_Clean": ",".join(sorted(skill_set)) if skill_set else "",
                "LLM_Total_Score": parsed.get("Total_Score") if parsed else None,
                "LLM_Shortlist": parsed.get("Shortlist") if parsed else None,
                "API_Error": api_error
            })
            processed.append(result_row)
        except Exception as e_row:
            # skip row but record minimal info
            processed.append({"summary": f"row_error: {e_row}", "shortlist": "No"})
            continue

    final_df = pd.DataFrame(processed)

    # standardize names for chart functions
    # Use capitals/expected keys
    # create fallback keys if needed
    if 'attrition_risk' in final_df.columns:
        final_df.rename(columns={'attrition_risk':'Attrition_Risk'}, inplace=True)
    if 'qualification_status' in final_df.columns:
        final_df.rename(columns={'qualification_status':'Over_Under_Qualified'}, inplace=True)
    if 'shortlist' in final_df.columns:
        final_df.rename(columns={'shortlist':'Final_Shortlist_Decision'}, inplace=True)
    if 'Education_Category' not in final_df.columns and 'education' in final_df.columns:
        final_df['Education_Category'] = final_df['education']

    # preview
    preview_df = final_df.head(MAX_PREVIEW_ROWS)

    # ---------- Charts generation ----------
    charts = {}
    insights = {}

    # Education Levels
    if 'Education_Category' in final_df.columns:
        s = final_df['Education_Category'].fillna('Unknown').astype(str)
        charts['education_pie'] = pie_from_series(s.value_counts(), "Education Levels Distribution")
        insights['education_pie'] = f"{s.value_counts().idxmax()} is the largest group ({s.value_counts().max()} candidates)."
    else:
        charts['education_pie'] = go.Figure().add_annotation(text="No education data", showarrow=False)
        insights['education_pie'] = "Education column missing."

    # Experience ranges (use Experience_Years_Clean)
    if 'Experience_Years_Clean' in final_df.columns:
        exp_series = pd.to_numeric(final_df['Experience_Years_Clean'], errors='coerce').fillna(-1)
        bins = [-1,2,5,10,999]
        labels = ['0-2 years','3-5 years','6-10 years','10+ years']
        exp_cat = pd.cut(exp_series, bins=bins, labels=labels)
        counts = exp_cat.value_counts().reindex(labels).fillna(0)
        charts['experience_bar'] = bar_from_counts(counts, "Experience Ranges (years)", x_title="Experience Range")
        insights['experience_bar'] = f"{counts.idxmax()} has the most candidates ({int(counts.max())})."
    else:
        charts['experience_bar'] = go.Figure().add_annotation(text="No experience data", showarrow=False)
        insights['experience_bar'] = "Experience data missing."

    # Attrition donut
    if 'Attrition_Risk' in final_df.columns:
        charts['attrition_donut'] = pie_from_series(final_df['Attrition_Risk'].fillna('Unknown').value_counts(), "Attrition Risk Segmentation")
        ins = final_df['Attrition_Risk'].fillna('Unknown').value_counts().to_dict()
        insights['attrition_donut'] = ", ".join([f"{k}: {v}" for k,v in ins.items()])
    else:
        charts['attrition_donut'] = go.Figure().add_annotation(text="No attrition data", showarrow=False)
        insights['attrition_donut'] = "Attrition risk missing."

    # Qualification status (Over/Good/Under)
    if 'Over_Under_Qualified' in final_df.columns:
        counts = final_df['Over_Under_Qualified'].fillna('Unknown').value_counts()
        charts['qualification_bar'] = bar_from_counts(counts, "Qualification Status")
        insights['qualification_bar'] = f"Counts: {counts.to_dict()}"
    else:
        charts['qualification_bar'] = go.Figure().add_annotation(text="No qualification data", showarrow=False)
        insights['qualification_bar'] = "Qualification data missing."

    # Education-to-role fit: if Education_to_Role_Fit exists or compute simple mapping
    if 'Education_to_Role_Fit' not in final_df.columns:
        # try compute basic fit using Education_Category vs jd_degree
        try:
            final_df['Education_to_Role_Fit'] = final_df['Education_Category'].apply(lambda e: ("Meets" if normalize_degree(str(e)) == jd_degree else ("Overqualified" if normalize_degree(str(e)) and DEGREE_ORDER.index(normalize_degree(str(e)))>DEGREE_ORDER.index(jd_degree) else ("Underqualified" if normalize_degree(str(e)) and DEGREE_ORDER.index(normalize_degree(str(e)))<DEGREE_ORDER.index(jd_degree) else "Unknown"))))
        except Exception:
            final_df['Education_to_Role_Fit'] = "Unknown"
    counts = final_df['Education_to_Role_Fit'].fillna('Unknown').value_counts()
    charts['edu_role_fit'] = bar_from_counts(counts, "Education-to-Role Fit")
    insights['edu_role_fit'] = f"Counts: {counts.to_dict()}"

    # Top skills chart
    skills_col_choice = find_col(final_df, ['skills','skills_set_clean','skills_set','skills_set_clean'])
    if skills_col_choice and skills_col_choice in final_df.columns:
        # parse cleaned skills if present as comma separated
        skill_list = []
        for val in final_df[skills_col_choice].fillna(''):
            if isinstance(val, str) and val.strip():
                parts = re.split(r'[;,/|\n]+', val.lower())
                for p in parts:
                    t = p.strip()
                    if t:
                        skill_list.append(t)
        if skill_list:
            top = Counter(skill_list).most_common(15)
            skills_df = pd.DataFrame(top, columns=['skill','count']).sort_values('count', ascending=True)
            fig_sk = px.bar(skills_df, x='count', y='skill', orientation='h', title='Top Skills')
            charts['top_skills'] = fig_sk
            insights['top_skills'] = ", ".join([f"{s}({c})" for s,c in top[:5]])
        else:
            charts['top_skills'] = go.Figure().add_annotation(text="No skills listed", showarrow=False)
            insights['top_skills'] = "No skills data found."
    else:
        charts['top_skills'] = go.Figure().add_annotation(text="No skills column found", showarrow=False)
        insights['top_skills'] = "Skills column missing."

    # Shortlist funnel
    total = len(final_df)
    final_short = final_df[final_df.get('Final_Shortlist_Decision','No')=='Yes'].shape[0]
    funnel_df = pd.DataFrame({'stage':['Applied','Shortlisted','Rejected'], 'count':[total, final_short, total-final_short]})
    charts['funnel'] = px.funnel(funnel_df, x='count', y='stage', title='Shortlisting Funnel')
    insights['funnel'] = f"{final_short}/{total} shortlisted."

    # Shortlist by education
    if 'Education_Category' in final_df.columns:
        grp = final_df.groupby([final_df['Education_Category'].fillna('Unknown'), final_df.get('Final_Shortlist_Decision','No')]).size().reset_index(name='count')
        if not grp.empty:
            fig_counts = px.bar(grp, x='Education_Category', y='count', color='Final_Shortlist_Decision', barmode='group', title='Shortlist by Education (counts)')
            charts['shortlist_by_edu_counts'] = fig_counts
            # percent chart
            totals = grp.groupby('Education_Category')['count'].sum().reset_index(name='total')
            merged = grp.merge(totals, on='Education_Category')
            merged['pct'] = merged['count'] / merged['total'] * 100
            fig_pct = px.bar(merged[merged['Final_Shortlist_Decision']=='Yes'], x='Education_Category', y='pct', title='Shortlist % by Education')
            charts['shortlist_by_edu_pct'] = fig_pct
            insights['shortlist_by_edu'] = "Shortlist percent by education shown."
        else:
            charts['shortlist_by_edu_counts'] = go.Figure().add_annotation(text="No data", showarrow=False)
            charts['shortlist_by_edu_pct'] = go.Figure().add_annotation(text="No data", showarrow=False)
            insights['shortlist_by_edu'] = "No data."
    else:
        charts['shortlist_by_edu_counts'] = go.Figure().add_annotation(text="No education column", showarrow=False)
        charts['shortlist_by_edu_pct'] = go.Figure().add_annotation(text="No education column", showarrow=False)
        insights['shortlist_by_edu'] = "Education column missing."

    # Build final summary text
    try:
        total_candidates = len(final_df)
        num_shortlisted = final_df[final_df.get('Final_Shortlist_Decision','No')=='Yes'].shape[0]
        pct_short = (num_shortlisted/total_candidates*100) if total_candidates>0 else 0
        num_over = final_df[final_df.get('Over_Under_Qualified','')=='Overqualified'].shape[0]
        num_high_risk = final_df[final_df.get('Attrition_Risk','')=='High'].shape[0]
        top_sk_text = insights.get('top_skills','No skills info')
        summary_text = (f"Processed {total_candidates} candidates. Shortlisted: {num_shortlisted} ({pct_short:.1f}%). "
                        f"Overqualified: {num_over}. High attrition risk: {num_high_risk}. Top skills: {top_sk_text}")
    except Exception as e:
        summary_text = f"Summary generation error: {e}"

    # Return values: DataFrame preview, summary, then a set of chart objects and insight strings
    # We'll return a fixed ordering so Gradio outputs are stable
    outputs = [
        preview_df, summary_text,
        charts['education_pie'], insights['education_pie'],
        charts['experience_bar'], insights['experience_bar'],
        charts['attrition_donut'], insights['attrition_donut'],
        charts['qualification_bar'], insights['qualification_bar'],
        charts['edu_role_fit'], insights['edu_role_fit'],
        charts['top_skills'], insights['top_skills'],
        charts['funnel'], insights['funnel'],
        charts['shortlist_by_edu_counts'], insights['shortlist_by_edu'],
        charts['shortlist_by_edu_pct'], insights['shortlist_by_edu']
    ]

    return outputs

# ---------- Gradio UI ----------

# Build inputs and outputs
company_input = gr.Textbox(label="Company Name", lines=1, placeholder="Acme Corp")
jd_input = gr.Textbox(label="Job Description", lines=8, placeholder="Paste the JD here...")
file_input = gr.File(label="Resume dataset (CSV or Excel)")

# Build outputs (must match the ordering above)
out_df = gr.Dataframe(label="Phase 1: Processed Data (preview)")
out_summary = gr.Textbox(label="Key Findings Summary", lines=4)
# Pairs of chart + insight (chart: gr.Plot, insight: gr.Textbox)
out_edu_chart = gr.Plot(label="Education Levels")
out_edu_insight = gr.Textbox(label="Education Insight", interactive=False)
out_exp_chart = gr.Plot(label="Experience Ranges")
out_exp_insight = gr.Textbox(label="Experience Insight", interactive=False)
out_attr_chart = gr.Plot(label="Attrition Risk")
out_attr_insight = gr.Textbox(label="Attrition Insight", interactive=False)
out_qual_chart = gr.Plot(label="Qualification Status")
out_qual_insight = gr.Textbox(label="Qualification Insight", interactive=False)
out_edurole_chart = gr.Plot(label="Education-to-Role Fit")
out_edurole_insight = gr.Textbox(label="Edu-Role Insight", interactive=False)
out_skills_chart = gr.Plot(label="Top Skills")
out_skills_insight = gr.Textbox(label="Top Skills Insight", interactive=False)
out_funnel_chart = gr.Plot(label="Shortlist Funnel")
out_funnel_insight = gr.Textbox(label="Funnel Insight", interactive=False)
out_shortlist_byedu_chart = gr.Plot(label="Shortlist by Education (counts)")
out_shortlist_byedu_insight = gr.Textbox(label="Shortlist by Education Insight", interactive=False)
out_shortlist_byedu_pct_chart = gr.Plot(label="Shortlist by Education (%)")
out_shortlist_byedu_pct_insight = gr.Textbox(label="Shortlist by Education (%) Insight", interactive=False)

all_outputs = [
    out_df, out_summary,
    out_edu_chart, out_edu_insight,
    out_exp_chart, out_exp_insight,
    out_attr_chart, out_attr_insight,
    out_qual_chart, out_qual_insight,
    out_edurole_chart, out_edurole_insight,
    out_skills_chart, out_skills_insight,
    out_funnel_chart, out_funnel_insight,
    out_shortlist_byedu_chart, out_shortlist_byedu_insight,
    out_shortlist_byedu_pct_chart, out_shortlist_byedu_pct_insight
]

iface = gr.Interface(
    fn=process_file_and_analyze,
    inputs=[company_input, jd_input, file_input],
    outputs=all_outputs,
    title="Resume Screener & Visualizer",
    description="Upload a resume dataset and a job description. Phase 1: preview enriched DataFrame. Phase 2: charts + insights.",
    allow_flagging="never"
)

# Launch - in Colab set share=True to get a public URL
iface.launch(share=True)
```

```python
# Streamlit launcher cell with updated app.py (adds improved Attrition Risk calculation)
import sys, subprocess, importlib.util, os, time, socket, textwrap
from IPython.display import IFrame, HTML, display

# -------------------- CONFIG --------------------
OVERWRITE_APP = True        # If True, will write the bundled app.py (replace if exists). Set False to keep existing app.py.
APP_FILE = "app.py"        # Streamlit app filename to create / run
PORT = 8501
USE_NGROK = False          # Set True if you want a public ngrok URL (requires NGROK_AUTH_TOKEN)
NGROK_AUTH_TOKEN = ""     # Paste your ngrok authtoken here if you want public tunneling; leave empty to skip or if USE_NGROK=False
STREAMLIT_TIMEOUT = 20     # seconds to wait for streamlit to start
# -------------------------------------------------

# -------------------- Helper: ensure packages --------------------
def ensure_pkg(pkg_name, import_name=None):
    if import_name is None:
        import_name = pkg_name
    spec = importlib.util.find_spec(import_name)
    if spec is None:
        print(f"Installing {pkg_name} ...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg_name], stdout=subprocess.DEVNULL)
    else:
        pass

# ensure required packages
ensure_pkg("streamlit")
ensure_pkg("pyngrok")
ensure_pkg("psutil")
ensure_pkg("plotly")
ensure_pkg("pandas")
ensure_pkg("numpy")
ensure_pkg("python-dateutil")

# import after ensuring
import psutil
from pyngrok import ngrok

# -------------------- Create a Streamlit app file (if requested) --------------------
app_code = r'''
# app.py - Resume Screener (minimal expanded version) with improved Attrition Risk
import streamlit as st
import pandas as pd
import numpy as np
import re
from collections import Counter
import plotly.express as px
from dateutil import parser
from datetime import datetime

# Layout
st.set_page_config(page_title="Resume Screener", layout="wide")
st.title("Resume Screener — Demo")

st.markdown("Upload a resume dataset (CSV/XLSX) and paste a Job Description to analyze candidates. This demo computes Overqualified, Education-to-Role Fit, Attrition Risk, and a simple shortlist heuristic.")

col1, col2 = st.columns([2,1])
with col1:
    uploaded = st.file_uploader("Upload resume dataset (CSV or Excel)", type=['csv','xlsx','xls'])
    jd_text = st.text_area("Job Description (paste full JD here)", height=200)
with col2:
    st.write("Options")
    preview_n = st.number_input("Rows to process (preview)", min_value=10, max_value=10000, value=500, step=10)

# Basic utilities (small subset of previous functions)
DEGREE_ORDER = ["highschool","diploma","bachelor","master","phd"]
DEGREE_MAP = {
    r"(high\s*school|12th|hsc|ssc|secondary)":"highschool",
    r"(diploma|polytechnic)":"diploma",
    r"(bachelor|b\.?tech|bsc|ba|bcom|bca)":"bachelor",
    r"(master|m\.?tech|msc|ma|mba|mca)":"master",
    r"(ph\.?d|doctorate)":"phd"
}
ROLE_ORDER = ["intern","junior","mid","senior","lead","manager","director","executive"]
ROLE_HINTS = {
    "intern":"intern|trainee",
    "junior":"junior|\\bjr\\b|associate",
    "mid":"analyst|engineer|consultant|specialist|associate",
    "senior":"senior|\\bsr\\b|staff|principal",
    "lead":"\\blead\\b|team lead|tech lead",
    "manager":"manager|\\bmgr\\b|head of",
    "director":"director|head\\b",
    "executive":"\\bvp\\b|vice president|chief|cxo|cto|cfo|ceo"
}

def normalize_degree(text):
    if not isinstance(text,str): return None
    t = text.lower()
    for pat,val in DEGREE_MAP.items():
        if re.search(pat, t): return val
    if "ph" in t and "d" in t: return "phd"
    if "mba" in t or "master" in t: return "master"
    if "bachelor" in t or "b." in t: return "bachelor"
    if "diploma" in t: return "diploma"
    if "school" in t: return "highschool"
    return None

def infer_role_level(title):
    if not isinstance(title,str): return None
    t = title.lower()
    for lvl in ROLE_ORDER[::-1]:
        pat = ROLE_HINTS.get(lvl)
        if pat and re.search(pat, t): return lvl
    if "senior" in t or " sr " in t: return "senior"
    if "manager" in t: return "manager"
    return "mid"

def extract_jd_requirements(text):
    if not isinstance(text,str): return 2,4,"bachelor","junior", set()
    txt = text.lower()
    m = re.search(r'(\d{1,2})\s*(?:[-–to]{1,3})\s*(\d{1,2})\s*years', txt)
    if m:
        jd_min, jd_max = int(m.group(1)), int(m.group(2))
    else:
        m2 = re.search(r'(\d{1,2})\s*\+\s*years', txt)
        if m2:
            jd_min = int(m2.group(1)); jd_max = jd_min + 2
        else:
            jd_min, jd_max = 2,4
    jd_degree = "bachelor"
    for pat,deg in DEGREE_MAP.items():
        if re.search(pat, txt):
            jd_degree = deg
            break
    jd_role = "junior"
    for lvl in ROLE_ORDER[::-1]:
        if re.search(r'\b'+re.escape(lvl)+r'\b', txt):
            jd_role = lvl
            break
    skills = set(re.findall(r'\b([a-z0-9\+\#]+)\b', txt))
    return jd_min, jd_max, jd_degree, jd_role, skills

def compute_overqualification(exp_years, role_level, edu_level, jd_min, jd_max, jd_role_idx, jd_edu_idx):
    reasons=[]
    over=False
    try:
        if exp_years is not None:
            ev=float(exp_years)
            if jd_max is not None and ev >= (jd_max+3):
                over=True; reasons.append(f"{ev:.1f} yrs vs JD {jd_min}-{jd_max}")
    except: pass
    try:
        if role_level is not None and jd_role_idx is not None:
            cand_idx = ROLE_ORDER.index(role_level) if role_level in ROLE_ORDER else None
            if cand_idx is not None and cand_idx - jd_role_idx >=2:
                over=True; reasons.append(f"title '{role_level}' above JD '{ROLE_ORDER[jd_role_idx]}'")
    except: pass
    try:
        if edu_level is not None and jd_edu_idx is not None:
            edu_idx = DEGREE_ORDER.index(edu_level) if edu_level in DEGREE_ORDER else None
            if edu_idx is not None and edu_idx - jd_edu_idx >=2:
                over=True; reasons.append(f"education '{edu_level}' >> JD '{DEGREE_ORDER[jd_edu_idx]}'")
    except: pass
    return over, "; ".join(reasons) if reasons else ""

def compute_education_fit(edu_level, jd_edu_idx):
    if jd_edu_idx is None: return "Unknown","No JD degree"
    if edu_level is None: return "Underqualified","No degree detected"
    edu_idx = DEGREE_ORDER.index(edu_level) if edu_level in DEGREE_ORDER else None
    if edu_idx is None: return "Unknown", f"Unrecognized {edu_level}"
    if edu_idx < jd_edu_idx: return "Underqualified", f"{edu_level} < required"
    if edu_idx == jd_edu_idx: return "Meets", f"{edu_level} meets requirement"
    return "Overqualified", f"{edu_level} > required"

# IMPROVED: Attrition risk helper that parses text and uses more sensitive thresholds
def compute_attrition_risk(row):
    """
    Improved attrition detection:
     - Looks for explicit numeric fields (job_change_rate_per_year, Avg Tenure Months, Num Jobs)
     - Attempts parsing of free-text 'Work Experience' or similar columns to extract date ranges
     - Falls back to compute job_change_rate = num_jobs / experience_years
     - Thresholds lowered so High/Medium appear when signals exist
    """
    def find_field(keys):
        for k in keys:
            if k in row and pd.notna(row[k]) and str(row[k]).strip()!="":
                return row[k]
        # try case-insensitive
        for col in row.index:
            for k in keys:
                if col.lower() == k.lower() and pd.notna(row[col]) and str(row[col]).strip()!="":
                    return row[col]
        return None

    # try explicit numeric fields
    job_change_rate = None
    val = find_field(['job_change_rate_per_year','Job Change Rate','job_change_rate','change_rate'])
    if val is not None:
        try:
            job_change_rate = float(re.sub(r'[^\d.]','', str(val)))
        except:
            job_change_rate = None

    avg_tenure_months = None
    val = find_field(['Avg Tenure Months','avg_tenure_months','average_tenure_months','avg_tenure'])
    if val is not None:
        try:
            avg_tenure_months = float(re.sub(r'[^\d.]','', str(val)))
        except:
            avg_tenure_months = None

    num_jobs = None
    val = find_field(['Num Jobs','Num_Jobs','num_jobs','Number of Jobs','job_count','num_job_changes','job changes'])
    if val is not None:
        try:
            num_jobs = int(re.sub(r'[^\d]','', str(val)))
        except:
            num_jobs = None

    exp_years = None
    val = find_field(['Experience (Years)','Experience','experience_years','yoe','years'])
    if val is not None:
        try:
            exp_years = float(re.sub(r'[^\d.]','', str(val)))
        except:
            exp_years = None

    # If we lack job counts or avg tenure, try to parse free-text employment/history columns
    if (avg_tenure_months is None or num_jobs is None):
        candidate_texts = []
        for col in row.index:
            if re.search(r'work|experience|employment|history|resume|job', str(col), flags=re.I):
                if pd.notna(row[col]) and str(row[col]).strip()!="":
                    candidate_texts.append(str(row[col]))
        combined = " | ".join(candidate_texts)
        if combined:
            # extract date ranges like "Jan 2018 - Jun 2020", "2018 to present", "2018-2020"
            ranges = re.findall(r'([A-Za-z]{3,9}\s*\d{4}|\d{4})\s*(?:[-–to]{1,3})\s*(present|current|[A-Za-z]{3,9}\s*\d{4}|\d{4})', combined, flags=re.I)
            tenures = []
            if ranges:
                for s_raw, e_raw in ranges:
                    try:
                        s = parser.parse(s_raw, fuzzy=True, default=datetime(1900,1,1))
                    except:
                        try:
                            s = parser.parse(re.sub(r'[^\d]','', s_raw)+"-01-01")
                        except:
                            s = None
                    if isinstance(e_raw, str) and e_raw.lower() in ('present','current'):
                        e = datetime.today()
                    else:
                        try:
                            e = parser.parse(e_raw, fuzzy=True, default=datetime(1900,1,1))
                        except:
                            try:
                                e = parser.parse(re.sub(r'[^\d]','', e_raw)+"-01-01")
                            except:
                                e = None
                    if s and e:
                        months = (e.year - s.year) * 12 + (e.month - s.month)
                        if months >= 0:
                            tenures.append(months)
            # fallback: count lines/companies as jobs if we have bullet separators
            if not tenures:
                possible_jobs = re.split(r'\n|\r|\||-{3,}', combined)
                job_lines = [p.strip() for p in possible_jobs if len(p.strip())>20]  # heuristic threshold
                if len(job_lines) >= 2:
                    num_jobs = num_jobs or len(job_lines)
            else:
                num_jobs = num_jobs or len(tenures)
                avg_tenure_months = avg_tenure_months or (sum(tenures)/len(tenures) if len(tenures)>0 else None)

    # compute job_change_rate fallback
    if job_change_rate is None and num_jobs is not None and exp_years not in (None,0):
        try:
            job_change_rate = float(num_jobs) / float(exp_years)
        except:
            job_change_rate = None

    # pct short jobs (simple flag) if avg tenure known
    pct_short_jobs = None
    if avg_tenure_months is not None:
        pct_short_jobs = 1.0 if avg_tenure_months < 12 else 0.0

    # salary heuristic
    salary_val = None
    sal = find_field(['Salary Expectation ($)','salary_expectation','$salary','Expected Salary','Salary'])
    try:
        if sal is not None and str(sal).strip()!="":
            salary_val = float(re.sub(r'[^\d.]','', str(sal)))
    except:
        salary_val = None

    # Decision logic (more sensitive thresholds)
    reasons = []
    risk = "Low"
    try:
        # High signals (use lowered thresholds so some rows become High)
        if (job_change_rate is not None and job_change_rate >= 0.6) or \
           (avg_tenure_months is not None and avg_tenure_months < 12) or \
           (num_jobs is not None and num_jobs >= 4) or \
           (pct_short_jobs is not None and pct_short_jobs > 0.5):
            risk = "High"
            if job_change_rate is not None: reasons.append(f"job_change_rate {job_change_rate:.2f}/yr")
            if avg_tenure_months is not None: reasons.append(f"avg tenure {avg_tenure_months:.1f} months")
            if num_jobs is not None: reasons.append(f"{num_jobs} jobs")
            if pct_short_jobs is not None: reasons.append(f"pct_short_jobs {pct_short_jobs:.2f}")
        # Medium signals
        elif (job_change_rate is not None and job_change_rate >= 0.3) or \
             (avg_tenure_months is not None and 12 <= avg_tenure_months < 24) or \
             (num_jobs is not None and num_jobs == 3):
            risk = "Medium"
            if job_change_rate is not None: reasons.append(f"job_change_rate {job_change_rate:.2f}/yr")
            if avg_tenure_months is not None: reasons.append(f"avg tenure {avg_tenure_months:.1f} months")
            if num_jobs is not None: reasons.append(f"{num_jobs} jobs")
        else:
            risk = "Low"
            reasons.append("stable by heuristics")

        # salary bump: high salary + low experience increases risk
        if salary_val is not None and exp_years is not None:
            if salary_val > 120000 and exp_years < 5:
                if risk == "Low":
                    risk = "Medium"; reasons.append(f"high salary ${salary_val:,.0f} for {exp_years} yrs exp")
                elif risk == "Medium":
                    risk = "High"; reasons.append(f"very high salary ${salary_val:,.0f} for {exp_years} yrs exp")
    except Exception as e:
        risk = "Unknown"
        reasons.append(f"error computing risk: {e}")

    return risk, "; ".join(reasons)

# UI behavior
if uploaded is None:
    st.info("Upload a resume CSV/XLSX to get started.")
else:
    try:
        if uploaded.name.lower().endswith('.csv'):
            df = pd.read_csv(uploaded)
        else:
            df = pd.read_excel(uploaded)
    except Exception as e:
        st.error(f"Failed to read file: {e}")
        st.stop()

    st.success(f"Loaded {len(df)} rows.")
    st.write("Detected columns:", df.columns.tolist())

    jd_min, jd_max, jd_degree, jd_role, jd_skills = extract_jd_requirements(jd_text or "")
    jd_role_idx = ROLE_ORDER.index(jd_role) if jd_role in ROLE_ORDER else None
    jd_edu_idx = DEGREE_ORDER.index(jd_degree) if jd_degree in DEGREE_ORDER else None

    # Find likely column names
    def find_col(df, candidates):
        for c in candidates:
            if c in df.columns: return c
        lower_map = {col.lower():col for col in df.columns}
        for c in candidates:
            if c.lower() in lower_map: return lower_map[c.lower()]
        return None

    name_col = find_col(df, ['Name','Candidate','Full Name'])
    title_col = find_col(df, ['Job Role','Title','Designation','role'])
    edu_col = find_col(df, ['Education','Highest Education','Qualification','Degree'])
    exp_col = find_col(df, ['Experience (Years)','Experience','Years of Experience','yoe'])
    skills_col = find_col(df, ['Skills','Skillset','Key Skills','technical_skills'])
    projects_col = find_col(df, ['Projects Count','Projects','projects_count'])

    results=[]
    rows = min(len(df), int(preview_n))
    for i, row in df.head(rows).iterrows():
        title = row.get(title_col) if title_col else None
        edu_raw = row.get(edu_col) if edu_col else None
        exp_raw = row.get(exp_col) if exp_col else None
        skills_raw = row.get(skills_col) if skills_col else ""
        proj_raw = row.get(projects_col) if projects_col else None

        edu_norm = normalize_degree(str(edu_raw)) if pd.notna(edu_raw) else None
        role_level = infer_role_level(str(title)) if pd.notna(title) else None
        try:
            exp_val = float(re.sub(r'[^\d\.]','', str(exp_raw))) if pd.notna(exp_raw) and str(exp_raw).strip()!="" else None
        except:
            exp_val = None
        skill_set = set([p.strip() for p in re.split(r'[;,/|\n]+', str(skills_raw).lower()) if p.strip()])
        try:
            proj_val = int(re.sub(r'[^\d]','', str(proj_raw))) if pd.notna(proj_raw) and str(proj_raw).strip()!="" else 0
        except:
            proj_val = 0

        is_over, over_reason = compute_overqualification(exp_val, role_level, edu_norm, jd_min, jd_max, jd_role_idx, jd_edu_idx)
        edu_fit_label, edu_fit_reason = compute_education_fit(edu_norm, jd_edu_idx)

        skill_match_ratio = 0.0
        if jd_skills:
            skill_match_ratio = len(jd_skills & skill_set) / max(1, len(jd_skills))

        shortlist = "No"
        if not is_over:
            edu_ok = (edu_fit_label != "Underqualified")
            exp_ok = (exp_val is None) or (exp_val >= (jd_min - 1))
            skills_ok = (skill_match_ratio >= 0.4) or (len(jd_skills)==0)
            if edu_ok and exp_ok and skills_ok:
                shortlist = "Yes"

        # compute attrition risk for this row and add reasons
        attr_risk_label, attr_risk_reasons = compute_attrition_risk(row)

        out = row.to_dict()
        out.update({
            "summary": f"{int(skill_match_ratio*100)}% JD skills; exp={exp_val}",
            "attrition_risk": attr_risk_label,
            "attrition_reasons": attr_risk_reasons,
            "qualification_status": "Overqualified" if is_over else ("Good Fit" if edu_fit_label!="Underqualified" else "Underqualified"),
            "qualification_reason": over_reason or edu_fit_reason,
            "shortlist": shortlist,
            "Overqualified": "Yes" if is_over else "No",
            "Overqual_Reason": over_reason,
            "Edu_to_Role_Fit": edu_fit_label,
            "Edu_Fit_Reason": edu_fit_reason,
            "Skill_Match_Ratio": round(skill_match_ratio,3),
            "Projects_Count_Clean": proj_val
        })
        results.append(out)

    final_df = pd.DataFrame(results)
    st.subheader("Phase 1 — Enriched Data (preview)")
    st.dataframe(final_df, height=400)
    csv_bytes = final_df.to_csv(index=False).encode('utf-8')
    st.download_button("Download processed CSV (preview)", data=csv_bytes, file_name="processed_resume_preview.csv", mime="text/csv")

    st.subheader("Phase 2 — Charts & Key Findings")
    # education pie
    if 'Edu_to_Role_Fit' in final_df.columns:
        ec = final_df['Edu_to_Role_Fit'].fillna('Unknown').value_counts()
        fig = px.pie(values=ec.values, names=ec.index, title="Education-to-Role Fit", hole=0.35)
        st.plotly_chart(fig, use_container_width=True)
    # experience ranges
    if exp_col and exp_col in final_df.columns:
        exp_series = pd.to_numeric(final_df[exp_col], errors='coerce').fillna(-1)
        bins=[-1,2,5,10,999]
        labels=['0-2','3-5','6-10','10+']
        exp_cat = pd.cut(exp_series, bins=bins, labels=labels)
        counts = exp_cat.value_counts().reindex(labels).fillna(0)
        fig2 = px.bar(x=counts.index, y=counts.values, title="Experience Ranges")
        st.plotly_chart(fig2, use_container_width=True)
    # top skills
    if skills_col and skills_col in final_df.columns:
        all_sk = final_df[skills_col].fillna('').astype(str).tolist()
        skill_list=[]
        for s in all_sk:
            parts = re.split(r'[;,/|\n]+', s.lower())
            for p in parts:
                p=p.strip()
                if p:
                    skill_list.append(p)
        if skill_list:
            from collections import Counter
            top = Counter(skill_list).most_common(15)
            sk_names=[t[0] for t in top][::-1]
            sk_vals=[t[1] for t in top][::-1]
            fig3 = px.bar(x=sk_vals, y=sk_names, orientation='h', title="Top Skills (Top 15)")
            st.plotly_chart(fig3, use_container_width=True)

    # Attrition Risk pie
    if 'attrition_risk' in final_df.columns:
        ar = final_df['attrition_risk'].fillna('Unknown').value_counts()
        fig4 = px.pie(values=ar.values, names=ar.index, title="Attrition Risk Breakdown", hole=0.35)
        st.plotly_chart(fig4, use_container_width=True)

    st.markdown("---")
    st.write("Note: LLM integration is disabled in this demo. To enable LLM scoring, integrate your LLM client and add API handling.")
'''

if OVERWRITE_APP or (not os.path.exists(APP_FILE)):
    with open(APP_FILE, "w", encoding="utf-8") as f:
        f.write(app_code)
    print(f"Written {APP_FILE} (OVERWRITE_APP={OVERWRITE_APP}).")
else:
    print(f"{APP_FILE} exists and OVERWRITE_APP=False — leaving it untouched.")

# -------------------- Start Streamlit app as background process --------------------
import subprocess, psutil, signal

# kill any previous streamlit processes started with the same app (best-effort)
def kill_existing_streamlit(app_name=APP_FILE):
    for p in psutil.process_iter(['pid','name','cmdline']):
        try:
            cmd = p.info.get('cmdline') or []
            if cmd and 'streamlit' in " ".join(cmd) and app_name in " ".join(cmd):
                print("Killing previous process:", p.pid, cmd)
                p.kill()
        except Exception:
            pass

kill_existing_streamlit(APP_FILE)

STREAMLIT_CMD = [sys.executable, "-m", "streamlit", "run", APP_FILE, "--server.headless", "true", "--server.port", str(PORT)]
print("Launching Streamlit with command:", " ".join(STREAMLIT_CMD))
proc = subprocess.Popen(STREAMLIT_CMD, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, preexec_fn=os.setsid if os.name!="nt" else None)

# wait for port to be open
def wait_for_port(port, timeout=STREAMLIT_TIMEOUT):
    start = time.time()
    while time.time() - start < timeout:
        s = socket.socket()
        try:
            s.connect(("127.0.0.1", port)); s.close(); return True
        except Exception:
            time.sleep(0.5)
    return False

if not wait_for_port(PORT, timeout=STREAMLIT_TIMEOUT):
    print(f"Warning: Streamlit did not open port {PORT} in {STREAMLIT_TIMEOUT}s. Showing recent stderr:")
    try:
        err = proc.stderr.read()
        print(err[-2000:])
    except Exception as e:
        print("Could not read stderr:", e)

# -------------------- Create ngrok tunnel if requested --------------------
public_url = None
if USE_NGROK:
    if not NGROK_AUTH_TOKEN:
        print("NGROK_AUTH_TOKEN empty — cannot create ngrok tunnel. Set NGROK_AUTH_TOKEN and rerun or set USE_NGROK=False.")
    else:
        try:
            ngrok.set_auth_token(NGROK_AUTH_TOKEN)
            tunnel = ngrok.connect(addr=PORT, proto="http")
            public_url = str(tunnel)
            print("Ngrok public URL:", public_url)
        except Exception as e:
            print("Ngrok error:", e)
            public_url = None

# -------------------- Embed app in notebook --------------------
if public_url:
    url = public_url if public_url.startswith("http") else f"http://{public_url}"
    display(HTML(f'<b>Streamlit app (ngrok):</b> <a href="{url}" target="_blank">{url}</a>'))
    display(IFrame(src=url, width="100%", height=700))
else:
    local_url = f"http://localhost:{PORT}"
    print("Opening local URL:", local_url)
    display(HTML(f'<b>Streamlit app (local):</b> <a href="{local_url}" target="_blank">{local_url}</a>'))
    display(IFrame(src=local_url, width="100%", height=700))

# -------------------- Cleanup helper --------------------
def cleanup():
    # kill ngrok tunnels
    try:
        ngrok.kill()
    except Exception:
        pass
    # kill streamlit processes for this app
    for p in psutil.process_iter(['pid','name','cmdline']):
        try:
            cmd = p.info.get('cmdline') or []
            if cmd and 'streamlit' in " ".join(cmd) and APP_FILE in " ".join(cmd):
                print("Killing streamlit process", p.pid)
                p.kill()
        except Exception:
            pass
    print("Cleanup attempted. If streamlit process persists, restart kernel or kill manually.")
# expose cleanup to user namespace
globals()['cleanup'] = cleanup

print("Done — use cleanup() to stop the app and close tunnels.")

```
